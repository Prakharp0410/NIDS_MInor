```python
# ==============================================================================
# FROM-SCRATCH JOURNAL PIPELINE
# CNN-LSTM for NSL-KDD, CICIDS2017, and UNSW-NB15 (SEPARATE EXPERIMENTS)
# ==============================================================================

import os
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import mutual_info_classif
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix,
    classification_report, roc_curve
)
from sklearn.utils.class_weight import compute_class_weight

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, BatchNormalization, LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

# ------------------------------------------------------------------------------
# 1. RANDOM SEED
# ------------------------------------------------------------------------------

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# ------------------------------------------------------------------------------
# 2. KAGGLE PATHS
# ------------------------------------------------------------------------------

NSL_PATH  = "/kaggle/input/datasets/hassan06/nslkdd"
CIC_PATH  = "/kaggle/input/datasets/dhoogla/cicids2017"
UNSW_PATH = "/kaggle/input/datasets/dhoogla/unswnb15"

# ------------------------------------------------------------------------------
# 3. HELPER FUNCTIONS
# ------------------------------------------------------------------------------

def list_data_files(base_path):
    files = []
    valid_ext = (".csv", ".txt", ".parquet", ".feather")
    for root, _, filenames in os.walk(base_path):
        for f in filenames:
            if f.lower().endswith(valid_ext):
                files.append(os.path.join(root, f))
    return sorted(files)

def safe_read_file(path):
    ext = os.path.splitext(path)[1].lower()

    try:
        if ext == ".parquet":
            return pd.read_parquet(path)

        if ext == ".feather":
            return pd.read_feather(path)

        if ext in [".csv", ".txt"]:
            for sep in [",", ";", "\t"]:
                try:
                    df = pd.read_csv(path, sep=sep, low_memory=False)
                    if df.shape[1] >= 2:
                        return df
                except:
                    pass

            try:
                df = pd.read_csv(path, sep=r"\s+", engine="python", low_memory=False)
                if df.shape[1] >= 2:
                    return df
            except:
                pass

            for sep in [",", ";", "\t", r"\s+"]:
                try:
                    df = pd.read_csv(path, sep=sep, header=None, engine="python", low_memory=False)
                    if df.shape[1] >= 2:
                        return df
                except:
                    pass
    except Exception as e:
        print(f"Read failed for {path}: {e}")

    return None

def clean_columns(df):
    df.columns = [
        str(c).strip().lower().replace(" ", "_").replace("-", "_").replace("/", "_")
        for c in df.columns
    ]
    return df

def normalize_text(x):
    if pd.isna(x):
        return ""
    return str(x).strip().lower()

def normalize_binary_label(x):
    x = normalize_text(x)
    if x in {"normal", "normal.", "benign", "0"}:
        return 0
    if "normal" in x or "benign" in x:
        return 0
    return 1

def detect_label_column(df):
    candidates = ["binary_label", "label", "labels", "class", "attack_cat", "attack", "target", "outcome"]
    for c in candidates:
        if c in df.columns:
            return c
    return None

def balanced_sample(df, max_rows=50000, label_col="binary_label", random_state=42):
    if len(df) <= max_rows:
        return df.copy()

    if label_col not in df.columns:
        return df.sample(n=max_rows, random_state=random_state).reset_index(drop=True)

    normal = df[df[label_col] == 0]
    attack = df[df[label_col] == 1]

    if len(normal) == 0 or len(attack) == 0:
        return df.sample(n=max_rows, random_state=random_state).reset_index(drop=True)

    n_each = max_rows // 2
    normal_sample = normal.sample(min(len(normal), n_each), random_state=random_state)
    attack_sample = attack.sample(min(len(attack), n_each), random_state=random_state)

    sampled = pd.concat([normal_sample, attack_sample], axis=0)
    sampled = sampled.sample(frac=1, random_state=random_state).reset_index(drop=True)
    return sampled

# ------------------------------------------------------------------------------
# 4. DATASET LOADERS
# ------------------------------------------------------------------------------

def load_nsl_kdd(base_path):
    files = list_data_files(base_path)
    print(f"NSL-KDD files found: {len(files)}")

    dfs = []
    for fp in files:
        df = safe_read_file(fp)
        if df is None:
            continue

        if df.shape[1] in [42, 43]:
            nsl_cols = [
                "duration","protocol_type","service","flag","src_bytes","dst_bytes",
                "land","wrong_fragment","urgent","hot","num_failed_logins","logged_in",
                "num_compromised","root_shell","su_attempted","num_root","num_file_creations",
                "num_shells","num_access_files","num_outbound_cmds","is_host_login",
                "is_guest_login","count","srv_count","serror_rate","srv_serror_rate",
                "rerror_rate","srv_rerror_rate","same_srv_rate","diff_srv_rate",
                "srv_diff_host_rate","dst_host_count","dst_host_srv_count",
                "dst_host_same_srv_rate","dst_host_diff_srv_rate",
                "dst_host_same_src_port_rate","dst_host_srv_diff_host_rate",
                "dst_host_serror_rate","dst_host_srv_serror_rate",
                "dst_host_rerror_rate","dst_host_srv_rerror_rate","label"
            ]
            if df.shape[1] == 43:
                nsl_cols.append("difficulty")
            df.columns = nsl_cols[:df.shape[1]]
        else:
            df = clean_columns(df)

        df = clean_columns(df)
        label_col = detect_label_column(df)
        if label_col is None:
            continue

        df["binary_label"] = df[label_col].apply(normalize_binary_label)
        df["dataset_source"] = "NSL-KDD"
        dfs.append(df)

    if not dfs:
        raise ValueError("No NSL-KDD files could be loaded.")

    df = pd.concat(dfs, ignore_index=True)
    print("NSL-KDD shape:", df.shape)
    return df

def load_cicids2017(base_path):
    files = list_data_files(base_path)
    print(f"CICIDS2017 files found: {len(files)}")

    dfs = []
    for fp in files:
        df = safe_read_file(fp)
        if df is None:
            continue

        df = clean_columns(df)
        label_col = detect_label_column(df)
        if label_col is None:
            continue

        df["binary_label"] = df[label_col].apply(normalize_binary_label)
        df["dataset_source"] = "CICIDS2017"
        dfs.append(df)

    if not dfs:
        raise ValueError("No CICIDS2017 files could be loaded.")

    df = pd.concat(dfs, ignore_index=True)
    print("CICIDS2017 shape:", df.shape)
    return df

def load_unsw_nb15(base_path):
    files = list_data_files(base_path)
    print(f"UNSW-NB15 files found: {len(files)}")

    dfs = []
    for fp in files:
        df = safe_read_file(fp)
        if df is None:
            continue

        df = clean_columns(df)

        if "label" in df.columns:
            df["binary_label"] = pd.to_numeric(df["label"], errors="coerce").fillna(0).astype(int).clip(0, 1)
        else:
            label_col = detect_label_column(df)
            if label_col is None:
                continue
            df["binary_label"] = df[label_col].apply(normalize_binary_label)

        df["dataset_source"] = "UNSW-NB15"
        dfs.append(df)

    if not dfs:
        raise ValueError("No UNSW-NB15 files could be loaded.")

    df = pd.concat(dfs, ignore_index=True)
    print("UNSW-NB15 shape:", df.shape)
    return df

# ------------------------------------------------------------------------------
# 5. LOAD DATA
# ------------------------------------------------------------------------------

nsl_df  = load_nsl_kdd(NSL_PATH)
cic_df  = load_cicids2017(CIC_PATH)
unsw_df = load_unsw_nb15(UNSW_PATH)

# optional sampling for faster Kaggle runs
nsl_df  = balanced_sample(nsl_df,  max_rows=50000)
cic_df  = balanced_sample(cic_df,  max_rows=50000)
unsw_df = balanced_sample(unsw_df, max_rows=50000)

print("\nSampled shapes:")
print("NSL-KDD   :", nsl_df.shape)
print("CICIDS2017:", cic_df.shape)
print("UNSW-NB15 :", unsw_df.shape)

# ------------------------------------------------------------------------------
# 6. SINGLE-DATASET PIPELINE
# ------------------------------------------------------------------------------

def preprocess_single_dataset(df, dataset_name="Dataset", top_k_features=40):
    df = df.copy()

    if "binary_label" in df.columns:
        target_col = "binary_label"
    elif "label" in df.columns:
        target_col = "label"
    else:
        raise ValueError(f"{dataset_name}: no label column found.")

    if not pd.api.types.is_numeric_dtype(df[target_col]):
        df[target_col] = df[target_col].astype(str).str.lower().str.strip()
        df[target_col] = df[target_col].apply(normalize_binary_label)

    df[target_col] = pd.to_numeric(df[target_col], errors="coerce").fillna(0).astype(int).clip(0, 1)

    # drop leakage/meta fields
    drop_cols = [c for c in ["dataset_source"] if c in df.columns]
    df = df.drop(columns=drop_cols, errors="ignore")

    X = df.drop(columns=[target_col]).copy()
    y = df[target_col].copy()

    categorical_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()

    print("\n" + "-"*70)
    print(f"{dataset_name}")
    print("-"*70)
    print("Target column:", target_col)
    print("Categorical columns:", categorical_cols)
    print("Original feature shape:", X.shape)
    print("Target distribution:")
    print(y.value_counts())

    if len(categorical_cols) > 0:
        X = pd.get_dummies(X, columns=categorical_cols, drop_first=True)

    X = X.apply(pd.to_numeric, errors="coerce")
    X = X.replace([np.inf, -np.inf], np.nan).fillna(0)

    top_k = min(top_k_features, X.shape[1])
    mi = mutual_info_classif(X, y, random_state=SEED)
    mi_series = pd.Series(mi, index=X.columns).sort_values(ascending=False)
    selected_features = mi_series.head(top_k).index.tolist()
    X = X[selected_features]

    print("Selected feature shape:", X.shape)
    print("Top selected features:", selected_features[:15])

    return X, y, selected_features

def create_sequences(data, labels, timesteps=20, stride=5):
    X_seq, y_seq = [], []
    labels = np.asarray(labels)

    for i in range(0, len(data) - timesteps, stride):
        X_seq.append(data[i:i + timesteps])
        y_seq.append(labels[i + timesteps - 1])

    return np.array(X_seq, dtype=np.float32), np.array(y_seq, dtype=np.int32)

def build_cnn_lstm_model(timesteps, n_features):
    model = Sequential([
        Conv1D(128, 3, activation="relu", padding="same", input_shape=(timesteps, n_features)),
        BatchNormalization(),
        MaxPooling1D(2),

        Conv1D(256, 3, activation="relu", padding="same"),
        BatchNormalization(),
        MaxPooling1D(2),

        LSTM(128, return_sequences=True),
        Dropout(0.4),

        LSTM(64),
        Dropout(0.4),

        Dense(64, activation="relu"),
        Dense(1, activation="sigmoid")
    ])

    model.compile(
        optimizer="adam",
        loss="binary_crossentropy",
        metrics=["accuracy", tf.keras.metrics.AUC(name="auc")]
    )
    return model

def run_single_dataset_experiment(
    df,
    dataset_name="Dataset",
    timesteps=20,
    stride=5,
    top_k_features=40,
    batch_size=64,
    epochs=40,
    show_plots=True
):
    X, y, selected_features = preprocess_single_dataset(
        df, dataset_name=dataset_name, top_k_features=top_k_features
    )

    # split before scaling and sequencing
    X_train_df, X_temp_df, y_train_df, y_temp_df = train_test_split(
        X, y, test_size=0.30, random_state=SEED, stratify=y
    )

    X_val_df, X_test_df, y_val_df, y_test_df = train_test_split(
        X_temp_df, y_temp_df, test_size=0.50, random_state=SEED, stratify=y_temp_df
    )

    print("Row splits:")
    print("Train:", X_train_df.shape, y_train_df.shape)
    print("Val  :", X_val_df.shape, y_val_df.shape)
    print("Test :", X_test_df.shape, y_test_df.shape)

    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train_df)
    X_val_scaled   = scaler.transform(X_val_df)
    X_test_scaled  = scaler.transform(X_test_df)

    X_train, y_train = create_sequences(X_train_scaled, y_train_df.values, timesteps=timesteps, stride=stride)
    X_val, y_val     = create_sequences(X_val_scaled, y_val_df.values, timesteps=timesteps, stride=stride)
    X_test, y_test   = create_sequences(X_test_scaled, y_test_df.values, timesteps=timesteps, stride=stride)

    print("Sequence splits:")
    print("Train:", X_train.shape, y_train.shape)
    print("Val  :", X_val.shape, y_val.shape)
    print("Test :", X_test.shape, y_test.shape)

    print("Sequence label distribution:")
    print("Train:", np.bincount(y_train))
    print("Val  :", np.bincount(y_val))
    print("Test :", np.bincount(y_test))

    class_weights = compute_class_weight(
        class_weight="balanced",
        classes=np.unique(y_train),
        y=y_train
    )
    class_weight_dict = dict(zip(np.unique(y_train), class_weights))
    print("Class weights:", class_weight_dict)

    model = build_cnn_lstm_model(timesteps, X_train.shape[2])
    model.summary()

    early_stop = EarlyStopping(
        monitor="val_loss",
        patience=6,
        restore_best_weights=True,
        verbose=1
    )

    reduce_lr = ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=3,
        verbose=1
    )

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        class_weight=class_weight_dict,
        callbacks=[early_stop, reduce_lr],
        verbose=1
    )

    y_pred_prob = model.predict(X_test).ravel()
    y_pred = (y_pred_prob >= 0.5).astype(int)

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    roc_auc = roc_auc_score(y_test, y_pred_prob)

    tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
    fpr_value = fp / (fp + tn + 1e-12)

    print(f"\n{dataset_name} Evaluation Metrics")
    print(f"Accuracy : {accuracy:.6f}")
    print(f"Precision: {precision:.6f}")
    print(f"Recall   : {recall:.6f}")
    print(f"F1 Score : {f1:.6f}")
    print(f"ROC AUC  : {roc_auc:.6f}")
    print(f"FPR      : {fpr_value:.6f}")

    print("\nClassification Report")
    print(classification_report(y_test, y_pred, zero_division=0))

    if show_plots:
        plt.style.use("seaborn-v0_8")

        cm = confusion_matrix(y_test, y_pred)
        plt.figure(figsize=(6, 5))
        sns.heatmap(
            cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=["Normal", "Attack"],
            yticklabels=["Normal", "Attack"]
        )
        plt.title(f"{dataset_name} - Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()

        fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
        plt.figure(figsize=(6, 5))
        plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.4f}", linewidth=2)
        plt.plot([0, 1], [0, 1], "k--")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"{dataset_name} - ROC Curve")
        plt.legend()
        plt.show()

        plt.figure(figsize=(6, 5))
        plt.plot(history.history["accuracy"], label="Train")
        plt.plot(history.history["val_accuracy"], label="Validation")
        plt.title(f"{dataset_name} - Accuracy Curve")
        plt.xlabel("Epoch")
        plt.ylabel("Accuracy")
        plt.legend()
        plt.show()

        plt.figure(figsize=(6, 5))
        plt.plot(history.history["loss"], label="Train")
        plt.plot(history.history["val_loss"], label="Validation")
        plt.title(f"{dataset_name} - Loss Curve")
        plt.xlabel("Epoch")
        plt.ylabel("Loss")
        plt.legend()
        plt.show()

    results_row = {
        "Dataset": dataset_name,
        "Accuracy": accuracy,
        "Precision": precision,
        "Recall": recall,
        "F1 Score": f1,
        "ROC-AUC": roc_auc,
        "FPR": fpr_value,
        "Num_Features": X.shape[1],
        "Timesteps": timesteps,
        "Stride": stride
    }

    return results_row, model, history, selected_features

# ------------------------------------------------------------------------------
# 7. RUN EXPERIMENTS
# ------------------------------------------------------------------------------

all_results = []

res_nsl, model_nsl, hist_nsl, feat_nsl = run_single_dataset_experiment(
    nsl_df,
    dataset_name="NSL-KDD",
    timesteps=20,
    stride=5,
    top_k_features=40,
    batch_size=64,
    epochs=40,
    show_plots=True
)
all_results.append(res_nsl)

res_cic, model_cic, hist_cic, feat_cic = run_single_dataset_experiment(
    cic_df,
    dataset_name="CICIDS2017",
    timesteps=20,
    stride=5,
    top_k_features=40,
    batch_size=64,
    epochs=40,
    show_plots=True
)
all_results.append(res_cic)

res_unsw, model_unsw, hist_unsw, feat_unsw = run_single_dataset_experiment(
    unsw_df,
    dataset_name="UNSW-NB15",
    timesteps=20,
    stride=5,
    top_k_features=40,
    batch_size=64,
    epochs=40,
    show_plots=True
)
all_results.append(res_unsw)

# ------------------------------------------------------------------------------
# 8. FINAL RESULTS TABLE
# ------------------------------------------------------------------------------

results_df = pd.DataFrame(all_results)

for col in ["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC", "FPR"]:
    results_df[col] = results_df[col].round(4)

print("\nFINAL DATASET-WISE RESULTS TABLE")
print(results_df)

results_df.to_csv("/kaggle/working/dataset_wise_cnn_lstm_results.csv", index=False)
print("\nSaved: /kaggle/working/dataset_wise_cnn_lstm_results.csv")

# ------------------------------------------------------------------------------
# 9. BAR CHART FOR PAPER
# ------------------------------------------------------------------------------

sns.set_style("whitegrid")

plot_df = results_df.set_index("Dataset")[["Accuracy", "Precision", "Recall", "F1 Score", "ROC-AUC"]]

ax = plot_df.plot(kind="bar", figsize=(10, 6), width=0.75)
plt.title("Performance Comparison of CNN-LSTM Across IDS Datasets", fontsize=14)
plt.ylabel("Metric Score", fontsize=12)
plt.xlabel("Dataset", fontsize=12)
plt.ylim(0, 1.0)
plt.xticks(rotation=0)

for container in ax.containers:
    ax.bar_label(container, fmt="%.3f", fontsize=8, padding=2)

plt.legend(loc="upper right")
plt.tight_layout()
plt.savefig("/kaggle/working/dataset_comparison_bar_chart.png", dpi=300)
plt.show()

print("\nSaved: /kaggle/working/dataset_comparison_bar_chart.png")
```
